First define the commodity class, which has three attributes: id, name and price. Soon there will be a dictionary to store a number of instances of this class.

```python
class Commodity:
    """define the attributes of commodity"""

    def __init__(self, c_id, c_name, price):
        self.c_id = c_id
        self.c_name = c_name
        self.price = price

    def __str__(self):
        return 'id number:%s\t name:%s\t\t price:%d' % (self.c_id, self.c_name, self.price)
```

The following is the content of all the commodities in the shop and all the users' information. Both of them will be stored in dictionaries in class `Manager`.

Now define the manager class, which has two attributes: all_in_shop and usr.The former is a dictionary, which stores all the commodities in the shop.The latter is also a dictionary, which stores the user information.
#### To be noticed: **You must log in via an existing user with corresponding password, otherwise you will be rejected.** 
You have 3 chances to try. If you fail, the system will shut down.
The existing users and their passwords are:
* Tom       142857
* Neo       112358
* Jeff      132134
* Amy       662724
* admin     123456

```python
class Manager:
    """define the attributes and functions of manager"""

    def __init__(self, all_in_shop, usr):
        self.all_in_shop = all_in_shop  #all the commodities in the shop
        self.usr = usr  # all the users' information

    def login(self):
        """divide the users into two groups: administrator and customer"""
        #give the user 3 chances to log in the system
        name = ''
        attempt = 0
        while True:
            if attempt >= 3:
                print('You have tried too many times.')
                return
            print('Enter your name:', end=' ', flush=True)
            name = input()
            #check if the user is in the system
            if not name in self.usr.keys():
                attempt += 1
                print('Your name is not in the system, please retry.')
                continue
            attempt = 0
            print(name, flush=True)
            break
        while True:
            if attempt >= 3:
                print('You have tried too many times.')
                return
            print('Enter your password:', end=' ', flush=True)
            pwd = input()
            #check if the password is correct
            if pwd != self.usr[name]:
                attempt += 1
                print('Your password is not correct, please retry.')
                continue
            print(pwd, flush=True)
            break
        #if the user is the administrator, he/she can access 4 functions
        if name == 'admin' and pwd == '123456':
            print('Welcome, dear administrator.')
            info = 'Enter “1” to display product information. \nEnter“2” to add product. \nEnter “3” to delete product.\nEnter“4” to quit the system\n\n'
            print(info)
            while True:
                input_num = input()
                if input_num == '1':
                    self.show_info()
                    print(info, flush=True)
                elif input_num == '2':
                    self.add_item()
                    print(info, flush=True)
                elif input_num == '3':
                    self.delete_item()
                    print(info, flush=True)
                elif input_num == '4':
                    print('System shutdown...', flush=True)
                    print('Bye!', flush=True)
                    return
                else:
                    print('Number illegal! Please try again.')
        else:
            #if the user is a customer, he/she can only buy things
            message = '''
Welcome to SDM Market. If you want to buy something, just follow the instructions to choose items you need.
            
Goods are listed in the following content:
            '''
            print(message)
            print()
            print()
            self.show_info()
            while True:
                print('Now enter name of the item you want:', end=' ', flush=True)
                item = input()
                #check if the item exists
                if not self.check_exist_by_name(item):
                    print('ItemNotFound!')
                    continue
                print(item)
                print('Enter the amount you need:', end=' ', flush=True)
                amount = input()
                #check if the amount is legal
                if int(amount) <= 0:
                    print('error, retry')
                    continue
                print(amount)
                pay = (int(amount)) * self.find_by_name(item).price
                print('Succeeded! You have paid $: $%d' % pay)
                #ask if the customer wants to buy something else
                print('\nDo you want to buy something else? Enter “y” to continue, others to quit the system.')
                command = input()
                if command == 'y':
                    self.show_info()
                    continue
                else:
                    print('See you next time!')
                    return

    def check_exist_by_name(self, name):
        for value in self.all_in_shop.values():
            if value.c_name == name:
                return True
        return False

    def find_by_name(self, name):
        if not self.check_exist_by_name(name):
            return None
        for value in self.all_in_shop.values():
            if value.c_name == name:
                return value

    def show_info(self):
        print('shop information:')
        for key in self.all_in_shop.keys():
            print(self.all_in_shop[key])
        print('\n\n')

    def add_item(self):
        while True:
            print('Enter a new id number:', end=' ', flush=True)
            id_new = input()
            #check if the id has been used
            if id_new in self.all_in_shop.keys():
                print('Sorry, id has been used, please try again.')
                continue
            print(id_new, flush=True)
            break
        print('Enter the name:', end=' ', flush=True)
        name_new = input()
        print(name_new)
        print('Enter the price:', end=' ', flush=True)
        price_new = input()
        print(price_new)
        new_item = Commodity(id_new, name_new, int(price_new))
        self.all_in_shop[new_item.c_id] = new_item
        print()
        print('Item added successfully!\n\n')

    def delete_item(self):
        while True:
            print('Enter the id number of the item you want to delete:',flush=True)
            id_to_delete = input()
            #check if the id exists
            if not id_to_delete in self.all_in_shop.keys():
                print('Sorry,id has not been found, please retry.')
                continue
            name_deleted = self.all_in_shop.pop(id_to_delete)
            print('item "' + name_deleted.c_name + ' "has been deleted successfully!\n\n', flush=True)
            break
```
