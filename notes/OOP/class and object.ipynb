# classes and objects

### A programmer-defined type is also called a class.

```python
class Point:
    """coordinate in 2D space"""

Point
```

```python
obj = Point()
obj# the return value is the address of the object in memory
```

### attributes

```python
obj.x = 5# assign value to the attribute of the object
obj.y = 10


def print_point(p):
    print('(%g, %g)' % (p.x, p.y))


print_point(obj)


def distance_between_points(p1, p2):
    import math
    dis = '{:.2f}'.format(math.sqrt((p1.x-p2.x)**2 + (p1.y-p2.y)**2))
    return 'the distance between the two points is ' + (str(dis))


obj2 = Point()
obj2.x = 10
obj2.y = 20
print(distance_between_points(obj, obj2))
```

```python
class Rectangle:
    """represent a rectangle"""
box = Rectangle()
box.width = 100
box.height = 200
box.corner = Point()
box.corner.x = 0.0
box.corner.y = 0.0
box.corner.x += 10#attributes are mutable
```

### Instances as return values

```python
def center_of_a_reactangle(rect):
    p = Point()
    p.x = rect.corner.x + rect.width/2
    p.y = rect.corner.y + rect.height/2
    return p

center = center_of_a_reactangle(box)
print_point(center)
```

### copying

#### shallow copy

```python
import copy
r1 = Rectangle()
r1.width = 10
r1.height = 20
r2 = copy.copy(r1)
print(r1 is r2)
r2.width = 100
print(r1.width)
```

To copy the contents of an object, including any **references to embedded objects**; implemented by the copy function in the copy module.

#### deep copy

```python
r3 = copy.deepcopy(r1)
print(r1 is r3)
r3.width = 1000
print(r3.width is r1.width)
```

To copy the contents of an object as well as any embedded objects, and **any objects embedded in them, and so on**; implemented by the deepcopy function in the copy module.
